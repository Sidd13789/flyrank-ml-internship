## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

### ML Task Type

I selected the **Content Refresh** lane. I frame this as a **classification** problem because the goal is to identify whether a webpage is likely to need a content refresh based on its historical performance and content-related features. The model will help classify pages into categories such as likely to need refresh or not, allowing content teams to prioritize their work more effectively.


## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

### Target / Proxy

For this project, I will use **trend_direction** as a proxy target. Pages marked as **"down"** indicate declining performance and are strong candidates for a content refresh. The model will learn patterns from the available features to help identify pages that may require attention. This target is suitable for the starter dataset because it reflects the current performance trend of each page.


## 3. Success metric

*One metric you can defend. What number means 'good'?*

### Success Metric

The success of this project will be measured using **Precision** and **Recall**. Precision is important because it helps ensure that pages recommended for refresh actually need attention, reducing unnecessary work. Recall is also important because missing pages that are truly declining could lead to lost traffic and missed opportunities. Together, these metrics provide a balanced way to evaluate how well the model supports content refresh decisions.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


A fixed rule, such as refreshing every page older than 180 days or every page with low clicks, cannot capture the combined effect of multiple factors. A page's performance depends on search volume, impressions, clicks, engagement, content age, competition, and other features working together.
Machine learning can learn patterns from many features at the same time and identify pages that are more likely to need a content refresh. This provides better decision support than relying on a single rule and helps content teams prioritize their work more effectively.
