## My Rule

The baseline rule prioritizes pages that are:
- Older content
- Low CTR
- High impressions

Older pages with low CTR and higher impressions receive higher scores because refreshing them may improve performance.

### Reason Codes

- STALE_LOW_CTR
- STALE_HIGH_VOLUME
- LOW_CTR
- HIGH_PRIORITY Baseline Rule

This baseline rule identifies pages that should be refreshed.

The score is based on three signals:

1. Older content receives higher priority.
2. Pages with low CTR receive higher priority.
3. Pages with higher search volume receive higher priority because improvements may have larger impact.

Pages with higher scores are ranked first.

### Reason Codes

| Code | Meaning |
|------|---------|
| STALE_LOW_CTR | Old page with low CTR |
| STALE_HIGH_VOLUME | Old page with high impressions |
| LOW_CTR | CTR is below expected |
| HIGH_PRIORITY | Highest overall score |
